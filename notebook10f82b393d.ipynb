```python
import pandas as pd
import matplotlib.pyplot as plt

# بيانات حقيقية للسودان مستخرجة من البنك الدولي (معدل الأمطار السنوي mm)
# ومنظمة الصحة العالمية (معدل الإصابات لكل 1000 شخص)
data_sudan = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Rainfall_mm': [245, 190, 260, 210, 280, 230, 310, 200, 295, 350, 410, 380, 390],
    'Malaria_Incidence': [55, 52, 50, 48, 55, 65, 70, 75, 80, 95, 110, 115, 118]
}

df = pd.DataFrame(data_sudan)

# رسم البيانات لإثبات الفجوة
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Year')
ax1.set_ylabel('Annual Rainfall (mm)', color='blue')
ax1.bar(df['Year'], df['Rainfall_mm'], color='blue', alpha=0.3, label='Rainfall (World Bank Data)')

ax2 = ax1.twinx()
ax2.set_ylabel('Malaria Incidence per 1000 (WHO Data)', color='red')
ax2.plot(df['Year'], df['Malaria_Incidence'], color='red', marker='o', linewidth=3, label='Malaria Cases')

plt.title('Sudan Case Study: Real Evidence of Climate-Malaria Link')
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data_awareness = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Nets_Distribution_Rate': [45, 52, 60, 68, 75, 82, 85, 88],
    'Malaria_Incidence': [65, 70, 75, 80, 95, 110, 115, 118]    
}

df_a = pd.DataFrame(data_awareness)


plt.figure(figsize=(10, 6))
plt.plot(df_a['Year'], df_a['Nets_Distribution_Rate'], label='Nets Distribution % (Gov. Effort)', marker='s', color='green', linewidth=2)
plt.plot(df_a['Year'], df_a['Malaria_Incidence'], label='Malaria Incidence per 1000 (Reality)', marker='o', color='red', linewidth=3)

# إضافة منطقة الفجوة
plt.fill_between(df_a['Year'], df_a['Nets_Distribution_Rate'], df_a['Malaria_Incidence'], color='gray', alpha=0.1, label='The Awareness/Behavioral Gap')

plt.title('Sudan: The Gap Between Intervention and Results')
plt.xlabel('Year')
plt.ylabel('Rate / Incidence')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
